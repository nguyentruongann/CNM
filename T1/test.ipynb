```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import networkx as nx
process_graph = nx.DiGraph()
nodes = [
    "Mở ứng dụng Grab",
    "Chọn loại dịch vụ (Xe máy, Ô tô, Taxi)",
    "Nhập địa điểm đón và điểm đến",
    "Xác nhận giá và đặt xe",
    "Tài xế nhận đơn",
    "Tài xế đến địa điểm đón",
    "Khách hàng lên xe",
    "Tài xế chở đến điểm đến",
    "Khách hàng thanh toán",
    "Đánh giá dịch vụ"
]

edges = [
    (nodes[i], nodes[i + 1]) for i in range(len(nodes) - 1)
]
process_graph.add_edges_from(edges)
pos = nx.spring_layout(process_graph, seed=42)  
plt.figure(figsize=(12, 8), facecolor="white")  
ax = plt.gca()
ax.set_facecolor('white')
nx.draw_networkx_nodes(process_graph, pos, node_size=3000, node_color="lightblue")
nx.draw_networkx_edges(process_graph, pos, arrowstyle="->", arrowsize=20, edge_color="black")
nx.draw_networkx_labels(process_graph, pos, font_size=10, font_color="black")
patch = mpatches.Patch(color="lightblue", label="Quy trình đặt xe Grab")
plt.legend(handles=[patch], loc="upper left")
plt.title("Quy Trình Đặt Xe Ứng Dụng Grab", fontsize=14)
plt.axis("off") 
plt.show()
```
